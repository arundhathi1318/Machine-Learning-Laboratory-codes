# Objective: learn to create dataframe and apply join operations between dataframes

1)Concatenating

```python
# dataframe 1
import pandas as pd  
  
# assign data of lists.  
data = {'city': ['delhi', 'mumbai', 'agra', 'goa'], 'positive': [20, 21, 19, 18],'neagtive': [120, 121, 119, 18] }  
  
# Create DataFrame  
df1 = pd.DataFrame(data)  
```

```python
df1
```

```python
# dataframe 2
data = {'city': ['delhi', 'mumbai', 'agra', 'chennai'], 'positive': [10, 21, 39, 18],'neagtive': [12, 101, 129, 118] }  
  
# Create DataFrame  
df2 = pd.DataFrame(data) 
df2
```

```python
# concatenate: concatenate the two dataframes one below the other.
df3 = pd.concat([df1,df2])
```

```python
df3
```

We see that in above result, we did not get continuous indexes( 0,1,2,3,0,1,2,3) to make them continuous like 0,1,2,3,4,â€¦ we can write
ignore_index=True

```python
df3 = pd.concat([df1,df2], ignore_index=True)
```

```python
df3
```

Assignng keys to dataframes df1 and df2

```python
df3 = pd.concat([df1,df2], keys = ['first', 'second'])
```

```python
# if you want to combine the two data frames horizontally means one next to another, apply axis=1, 1 signify column
df3 = pd.concat([df1,df2], axis =1)
df3
```

# Append:
The concat method can combine data frames along either rows or columns, while the append method only combines data frames along rows

```python
# dataframe 1
import pandas as pd  
  
# assign data of lists.  
data = {'city': ['delhi', 'mumbai', 'agra'], 'positive': [20, 21, 19],'neagtive': [120, 121, 119] }  
  
# Create DataFrame  
df1 = pd.DataFrame(data) 
df1
```

```python
# dataframe 2
import pandas as pd  
  
# assign data of lists.  
data = {'city': ['delhi', 'mumbai', 'agra'], 'positive': [210, 211, 19],'neagtive': [12, 121, 109] }  
  
# Create DataFrame  
df2 = pd.DataFrame(data) 
df2
```

```python
df3 = df1._append(df2)
```

```python
df3
```

# Merge data frames:In merging, you can merge two data frames to form a single data frame. You can also decide which columns you want to make common. 

# merge: always combine based on a column and we have to specify it, some column should be same in both dataframes based on which we can combine

```python
df3 = df1.merge(df2, on = 'city')
df3
```

```python
# positive_x	neagtive_x	belongs to first dataframe and positive_y	neagtive_y belong to second dataframe
```

We can join the dataframes in different ways:
 1)inner join: only common data of the dataframes are outputted
 2)left join:That means we should get all records of left dataframe and only the matching data of right dataframe.
 3)Right join:That means we should get all records of right dataframe and only the matching data of left dataframe.
 4)Full outer join: all data from right and left dataframe. if no matching NaN will come
    

```python
# inner join

df3 = df1.merge(df2, on = 'city', how ='inner')
df3
```

```python
# in the above output we cant see the change as all records were common, lets change df1 and add a new record
```

```python
# dataframe 1
import pandas as pd  
  
# assign data of lists.  
data = {'city': ['delhi', 'mumbai', 'agra', 'goa'], 'positive': [20, 21, 19, 88],'neagtive': [120, 121, 119, 133] }  
  
# Create DataFrame  
df1 = pd.DataFrame(data) 
df1
```

```python
# dataframe 2
import pandas as pd  
  
# assign data of lists.  
data = {'city': ['delhi', 'mumbai', 'agra'], 'positive': [210, 211, 19],'neagtive': [12, 121, 109] }  
  
# Create DataFrame  
df2 = pd.DataFrame(data) 
df2
```

```python
df3 = df1.merge(df2, on = 'city', how ='inner')
df3
```

```python
# we see that record for goa did not come as it was not common in both the dataframes
```

```python
# left join
df3 = df1.merge(df2, on = 'city', how ='left')
df3
```

```python
# Right join
df3 = df1.merge(df2, on = 'city', how ='right')
df3
```

```python
# outer join
# Right join
df3 = df1.merge(df2, on = 'city', how ='outer')
df3
```
